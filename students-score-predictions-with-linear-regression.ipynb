# Students scores predictions

**1. Importing the necessary libraries**

```python
import pandas as pd
import seaborn as sns
```

**2. Storing and visualizing the data**

```python
df = pd.read_csv('/kaggle/input/students-score-dataset-linear-regression/student_scores.csv')
```

```python
df.head() #first 5 rows
```

```python
df.shape #amount of rows & columns that the dataframe has
```

```python
df.info()
```

```python
df.isnull().sum() #we check if there is any null value
```

```python
df.plot(kind='scatter', x='Hours', y='Scores') #we plot the data to see the linear relationship between both variables
```

**3. Declaring de feature(X) and the target(Y)**

```python
feature = df[["Hours"]] #independent variable (X)
target = df["Scores"] #dependent variable (Y)
```

```python
feature.head()
```

**4. Random splitting of data**

```python
from sklearn.model_selection import train_test_split
```

*70% of data for training and 30% for testing*

```python
x_train, x_test, y_train, y_test  = train_test_split(feature,target, test_size=0.3, random_state=0)
```

**5. Creating the linear models**

```python
from sklearn.linear_model import LinearRegression 
```

*Instantiate the class*

```python
model_lr = LinearRegression()
```

**6. Fitting the linear model**

```python
model_lr.fit(x_train,y_train)
```

**7. Getting the score of the model**

*We multiply the score by 100 to get the percentage of goodness*

```python
lr_score = model_lr.score(x_test,y_test) * 100
```

```python
print("Linear regression model`s score --> ", int(lr_score),"%")
```

**8. Predicting "y" values for the testing explanatory variable `x_test`**

```python
y_pred = model_lr.predict(x_test)
```

**9. Compare the actual values (`y_test`) of the explanatory data (`x_test`) with their predicted values (`y_pred`)**

```python
y_pred
```

```python
import matplotlib.pyplot as plt
plt.title("Real test data VS Predicted score values")
plt.scatter(x_test, y_test, color ='b', label="Real test data") 
plt.plot(x_test, y_pred, color ='r', label="Predicted score values") 
plt.show()
```

**10. More info of the model**

```python
print("Coeficcient:",model_lr.coef_)
print("Intercept:",model_lr.intercept_)
```
